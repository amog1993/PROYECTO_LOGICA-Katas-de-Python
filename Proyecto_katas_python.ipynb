1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias
de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def funcion_diccionario (texto): 
    """
    Cuenta la frecuencia de cada carácter en un texto, excluyendo los espacios.

    Parámetros:
    texto (str): Cadena de texto a analizar.

    Retorna:
    dict[str, int]: Diccionario donde las claves son los caracteres y los valores son el número de veces que aparecen.
    """
    diccionario = {}
    for letra in texto: 
        if letra != " ":
            if letra in diccionario:
                diccionario[letra] +=1
            else: 
                diccionario.update({letra: 1})
    return diccionario

print(funcion_diccionario("Hola mundoooo"))
    
```

3. Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe
devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
def palabra_objetivo(lista, objetivo):
    """
    Filtra palabras de una lista que contienen una palabra objetivo.

    Parámetros:
    lista (list): Lista de palabras.
    objetivo (str): palabra objetivo

    Retorna:
    list: Lista con las palabras que contienen la palabra objetivo 
    """
    nueva_lista = []
    for palabra in lista:
        if objetivo in palabra:
            nueva_lista.append(palabra)
    return nueva_lista

print(palabra_objetivo(["hola", "adios", "holanda", "perro", "holandesa"], "hola"))
            
```

```python
#Definimos la función 
def funcion_1 (lista, nota_aprobada = 5):
    """
    Calcula el promedio de una lista de notas y determina si está aprobada.

    Parámetros:
    lista (list[int | float]): Lista de valores numéricos (notas).
    nota_aprobada (int | float, opcional): Nota mínima para aprobar. Por defecto es 5.

    Retorna:
    tuple:
        - float: Promedio de las notas.
        - str: "Aprobado" si el promedio es mayor o igual a la nota mínima,
               "Suspenso" en caso contrario.
    """
    suma = 0 
#Calcular suma de la lista
#Determinar si está aprobado o no 
    for numero in lista:
        suma += numero
    resultado = suma / len(lista)
    if resultado >= nota_aprobada:
        estado = "Aprobado"
    else: 
        estado = "Suspenso"

    return resultado, estado
#El ejercicio no pedía imprimir el resultado, pero podemos comprobamos que funciona 
#print(funcion_1([3,4,7], nota_aprobada= 5))
```

```python
def funcion_factorial(n):
    """
    Calcula el factorial de un número de forma recursiva.

    Parámetros:
    n (int): Número entero positivo.

    Retorna:
    int: Factorial de n.
    """
    if n == 1:
        return 1
    else:
        return n * funcion_factorial(n-1)

print(funcion_factorial(5)) 
```

10. Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una excepción personalizada y maneja el error adecuadamente.

```python
#Definimos la excepcion personalizada
class ListaVaciaError(Exception):
    pass

def calcular_promedio(lista):
    """
    Calcula el promedio de una lista de números.

    Parámetros:
    lista (list): Lista de números.

    Retorna:
    float: Promedio de los valores.

    Lanza:
    ListaVaciaError: Si la lista está vacía.
    """
    if not lista:
        raise ListaVaciaError("La lista está vacía")
    return sum(lista)/len(lista)

numeros_lista = [2,5,7,10,33]
#Manejamos el error
try:
    resultado = calcular_promedio(numeros_lista)
    print(f"El promedio de la lista es {resultado}")
except ListaVaciaError:
    print("Error: lista vacía")
```

30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras pero en diferente orden.

```python
def palabras_anagramas (a, b):
    """
    Determina si dos textos son anagramas.

    Compara dos cadenas ignorando mayúsculas, minúsculas y espacios. Verificando si contienen los mismos caracteres en diferente orden.

    Parámetros:
    a (str): Primera palabra o texto.
    b (str): Segunda palabra o texto.

    Retorna:
    str: "Palabras anagramas" si contienen los mismos caracteres,
         "Palabras no anagramas" en caso contrario.
    """
    a_sorted = sorted(a.lower().replace(" ", ""))
    b_sorted = sorted(b.lower().replace(" ", ""))

    if a_sorted == b_sorted: 
        return "Palabras anagramas"
    else: 
        return "Palabras no anagramas"
    
print(palabras_anagramas("Fresa", "fr ase"))
```

34. Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son: crecer_tronco , nueva_rama , crecer_ramas , quitar_rama e info_arbol . El objetivo es implementar estos métodos para manipular la estructura del árbol.

Código a seguir:

- Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
- Implementar el método crecer_tronco para aumentar la longitud del tronco en una unidad.
- Implementar el método nueva_rama para agregar una nueva rama de longitud 1 a la lista de ramas.
- Implementar el método crecer_ramas para aumentar en una unidad la longitud de todas las ramas existentes.
- Implementar el método quitar_rama para eliminar una rama en una posición específica.
- Implementar el método info_arbol para devolver información sobre la longitud del tronco, el número de ramas y las longitudes de las mismas.



```python
class Arbol:
    #Definitimos el metodo constructor 
    def __init__(self):
        """
        Inicializa un árbol con un tronco de longitud 1 y sin ramas.
        """
        self.tronco = 1
        self.ramas = []

#Definimos los métodos
    def crecer_tronco(self):
        """
        Incrementa en 1 la longitud del tronco.
        """
        self.tronco += 1
    
    def nueva_rama(self):
        """
        Añade una nueva rama al árbol con longitud inicial de 1.
        """
        self.ramas.append(1)

    def crecer_ramas(self):
        """
        Incrementa en 1 la longitud de todas las ramas del árbol.
        """
        for i in range(len(self.ramas)): 
            self.ramas[i] = self.ramas[i] + 1
    
    def quitar_rama(self, posicion):
        """
        Elimina una rama según su posición en la lista.

        Parámetros:
        posicion (int): Índice de la rama a eliminar.

        Retorna:
        None

        Nota:
        Si la posición no existe, muestra un mensaje por consola.
        """
        if posicion >= 0 and posicion < len(self.ramas):
            self.ramas.pop(posicion)
        else: 
            print("Posición no existe")

    def info_arbol(self):
        """
        Genera una descripción del estado actual del árbol.

        Incluye la longitud del tronco, el número de ramas y la longitud de cada una de ellas.

        Retorna:
        str: Descripción detallada del árbol.
        """
        detalle_ramas = ""
        for i, rama in enumerate(self.ramas):
            if i == 0:
                detalle_ramas += f"una mide {rama}"
            else:
                detalle_ramas += f" y otra mide {rama}"

        return f"El árbol tiene un tronco que mide {self.tronco}, tiene {len(self.ramas)} ramas: {detalle_ramas}."
```

```python
class UsuarioBanco:
    def __init__(self, nombre, saldo, cuenta_corriente):
        """
        Inicializa un usuario bancario.

        Parámetros:
        nombre (str): Nombre del titular de la cuenta.
        saldo (int | float): Saldo inicial disponible.
        cuenta_corriente (bool): Indica si el usuario tiene cuenta corriente.
        """
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    def retirar_dinero(self, cantidad):
        """
        Retira una cantidad de dinero del saldo si hay fondos suficientes.

        Parámetros:
        cantidad (int | float): Cantidad a retirar.

        Retorna:
        str: Mensaje indicando si el retiro fue exitoso o si hay saldo insuficiente.
        """
        if cantidad <=  self.saldo:
            self.saldo -= cantidad
            return f"Retiro Exitoso, su nuevo saldo es {self.saldo}"
        else: 
            return "Saldo insuficiente"

    def transferir_dinero(self, usuario2, cantidad):
        """
        Transfiere dinero desde otro usuario hacia este usuario.

        Parámetros:
        usuario2 (UsuarioBanco): Usuario desde el cual se retirará el dinero.
        cantidad (int | float): Cantidad a transferir.

        Retorna:
        str: Mensaje indicando si la transferencia fue exitosa o no.
        """
        resultado_retirar = usuario2.retirar_dinero(cantidad)
        if resultado_retirar == "Retiro Exitoso":
            self.saldo += cantidad
            return "Transferencia exitosa"
        else: 
            return "Saldo insuficiente"
        
    def agregar_dinero(self, cantidad):
        """
        Añade dinero al saldo del usuario.

        Parámetros:
        cantidad (int | float): Cantidad a ingresar.

        Retorna:
        str: Mensaje con el nuevo saldo actualizado.
        """
        self.saldo += cantidad
        return f"Dinero ingresado correctamente. Su nuevo saldo es {self.saldo}"
```

37. Crea una función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras , reemplazar_palabras , eliminar_palabra . Estas opciones son otras funciones que tenemos que definir primero y llamar dentro de la función procesar_texto.

    Código a seguir:
    
- Crear una función contar_palabras para contar el número de veces que aparece cada palabra en el texto. Tiene
que devolver un diccionario.
- Crear una función reemplazar_palabras para remplazar una palabra_original del texto por una palabra_nueva . Tiene
que devolver el texto con el remplazo de palabras.
- Crear una función eliminar_palabra para eliminar una palabra del texto. Tiene que devolver el texto con la palabra
eliminada.
- Crear la función procesar_texto que tome un texto, una opción(entre "contar", "reemplazar", "eliminar") y un
número de argumentos variable según la opción indicada.

```python
def contar_palabras(texto):
    """
    Cuenta la frecuencia de cada palabra en un texto.

    Parámetros:
    texto (str): Texto de entrada.

    Retorna:
    dict: Diccionario con palabras y su frecuencia.
    """
    texto = texto.lower()
    palabras1 = texto.split()
    diccionario = {}
    for palabra in palabras1:
        if palabra in diccionario:
            diccionario[palabra] += 1
        else: 
            diccionario[palabra] = 1 
    return diccionario

def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    """
    Reemplaza una palabra en un texto.

    Parámetros:
    texto (str): Texto original.
    palabra_original (str): Palabra a reemplazar.
    palabra_nueva (str): Nueva palabra.

    Retorna:
    str: Texto modificado.
    """
    reemplazo = texto.replace(palabra_original,  palabra_nueva)
    return reemplazo

def eliminar_palabra(texto, palabra):
    """
    Elimina una palabra de un texto.

    Parámetros:
    texto (str): Texto original.
    palabra (str): Palabra a eliminar.

    Retorna:
    str: Texto modificado.
    """
    quitar_palabra = texto.replace(palabra, "")
    return quitar_palabra

def procesar_texto(texto, opcion, *args):
    """
    Procesa un texto según una opción.

    Parámetros:
    texto (str): Texto de entrada.
    opcion (str): Tipo de operación ('contar', 'reemplazar', 'eliminar').
    "*"args: Argumentos adicionales según la operación.

    Retorna:
    any: Resultado de la operación seleccionada.
    """
    if opcion == "contar":
        return contar_palabras(texto)
    elif opcion == "reemplazar":
        return reemplazar_palabras(texto, args[0], args[1])
    elif opcion == "eliminar":
        return eliminar_palabra(texto, args[0])
    
```

40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o "triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura).

```python
import math 
def area_figura(figura, datos): 
    """
    Calcula el área de distintas figuras geométricas.

    Parámetros:
    figura (str): Tipo de figura ('triangulo', 'rectangulo', 'circulo').
    datos (tuple): Dimensiones necesarias.

    Retorna:
    str: Resultado con el área calculada.
    """
    if figura == "triangulo":
        area = (datos[0] * datos[1]) / 2
        return f" El area del triangulo es {area}"
    elif figura == "rectangulo":
        area = datos[0] * datos[1]
        return f" El area del rectangulo es {area}"
    elif figura == "circulo":
        area = math.pi * (datos[0] ** 2)
        return f" El area del circulo es {round(area, 2)}"
    else:
        return "Prueba con otra figura"
```
